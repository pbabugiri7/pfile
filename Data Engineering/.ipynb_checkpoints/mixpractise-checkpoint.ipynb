```python
# to swap the first and last number
p = []
n = int(input('Enter the no of objects assigning in list:'))
for i in range(0,n):
    c = int(input('Enter the element:' str(i+1) + ':'))
    p.append(c)
x = p[0]
p[0] = p[n-1]
p[n-1] = x
print(f'Swapping object and the new list is {p}')
```
